```python
client = MongoClient("mongodb://localhost:27017/")

# Selezione del database e della collezione
db = client["Flights"]
collection = db["Airlines"]

# Pipeline di aggregazione
pipeline = [
    {
        '$match': {
            'active': 'Y'
        }
    },
    {
        '$group': {
            '_id': '$country', 
            'total': {
                '$sum': 1
            }
        }
    },
    {
        "$sort":{
            "total":-1
        }
    },
    {
        "$limit":10
    },
    {
        "$out":"Top 10 country x Airlines"
    }
]

# Esecuzione della query
results = collection.aggregate(pipeline)
```

```python
db = client["Flights"]
collection = db["Top 10 country x Airlines"]

# Caricamento dei dati in un DataFrame
cursor = collection.find()
df = pd.DataFrame(list(cursor))
df.head()

sns.barplot(data=df, x='_id', y='total')

# 4. Mostra il grafico
plt.xticks(rotation=45, ha='right')  # opzionale, se i nomi sono lunghi
plt.title("Airlines x Country")
plt.figure(figsize=(20, 20)) 
plt.show()
```

```python
coll = client.Flights.Aereoporti

pipeline = [
    {
        '$group': {
            '_id': '$country',
            'airport_count': { '$sum': 1 }
        }
    },
    {
        "$sort":{
            "airport_count":-1
        }
    },
    {
        "$limit":10
    },
    {
        "$out":"Top 10 country x Aereoporti"
    }
    
]
    
results = coll.aggregate(pipeline)
```
